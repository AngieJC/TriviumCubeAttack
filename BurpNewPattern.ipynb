```python
def getTriviumIndex(x):
    if x < 288:
        return x
    xmod = x % 288
    if (xmod == 0) or (xmod == 93) or (xmod == 177):
        return x
    return getTriviumIndex(x - 288 - 1)

def generateTrivium(searchRound = 3, rootsIndex = [[840, 0]]):
    roots = []
    sys = {}
    for rootIndexNow in rootsIndex:
        roots.append(getTriviumIndex(rootIndexNow[0] * 288 + rootIndexNow[1]))
    for root in roots:
        sys[root] = []
    waiting = [[], []]
    for i in range(len(roots)):
        if rootsIndex[i][1] == 0:
            sys[roots[i]].append([getTriviumIndex(roots[i] - 288 + 68)])
            sys[roots[i]].append([getTriviumIndex(roots[i] - 288 + 242)])
            sys[roots[i]].append([getTriviumIndex(roots[i] - 288 + 287)])
            sys[roots[i]].append([getTriviumIndex(roots[i] - 288 + 285), getTriviumIndex(roots[i] - 288 + 286)])
            waiting[0].append(getTriviumIndex(roots[i] - 288 + 68))
            waiting[0].append(getTriviumIndex(roots[i] - 288 + 242))
            waiting[0].append(getTriviumIndex(roots[i] - 288 + 287))
            waiting[0].append(getTriviumIndex(roots[i] - 288 + 285))
            waiting[0].append(getTriviumIndex(roots[i] - 288 + 286))
        elif rootsIndex[i][1] == 93:
            sys[roots[i]].append([getTriviumIndex(roots[i] - 288 - 93 + 170)])
            sys[roots[i]].append([getTriviumIndex(roots[i] - 288 - 93 + 65)])
            sys[roots[i]].append([getTriviumIndex(roots[i] - 288 - 93 + 92)])
            sys[roots[i]].append([getTriviumIndex(roots[i] - 288 - 93 + 90), getTriviumIndex(roots[i] - 288 - 93 + 91)])
            waiting[0].append(getTriviumIndex(roots[i] - 288 - 93 + 170))
            waiting[0].append(getTriviumIndex(roots[i] - 288 - 93 + 65))
            waiting[0].append(getTriviumIndex(roots[i] - 288 - 93 + 92))
            waiting[0].append(getTriviumIndex(roots[i] - 288 - 93 + 90))
            waiting[0].append(getTriviumIndex(roots[i] - 288 - 93 + 91))
        elif rootsIndex[i][1] == 177:
            sys[roots[i]].append([getTriviumIndex(roots[i] - 288 - 177 + 263)])
            sys[roots[i]].append([getTriviumIndex(roots[i] - 288 - 177 + 161)])
            sys[roots[i]].append([getTriviumIndex(roots[i] - 288 - 177 + 176)])
            sys[roots[i]].append([getTriviumIndex(roots[i] - 288 - 177 + 174), getTriviumIndex(roots[i] - 288 - 177 + 175)])
            waiting[0].append(getTriviumIndex(roots[i] - 288 - 177 + 263))
            waiting[0].append(getTriviumIndex(roots[i] - 288 - 177 + 161))
            waiting[0].append(getTriviumIndex(roots[i] - 288 - 177 + 176))
            waiting[0].append(getTriviumIndex(roots[i] - 288 - 177 + 174))
            waiting[0].append(getTriviumIndex(roots[i] - 288 - 177 + 175))
    
    for i in range(searchRound - 1):
        waiting[0] = list(set(waiting[0]))
        waiting[1] = list(set(waiting[1]))
        while len(waiting[i % 2]):
            x = waiting[i % 2].pop()
            sys[x] = []
            if x % 288 == 0:
                sys[x].append([getTriviumIndex(x - 288 + 68)])
                sys[x].append([getTriviumIndex(x - 288 + 242)])
                sys[x].append([getTriviumIndex(x - 288 + 287)])
                sys[x].append([getTriviumIndex(x - 288 + 285), getTriviumIndex(x - 288 + 286)])
                waiting[(i + 1) % 2].append(getTriviumIndex(x - 288 + 68))
                waiting[(i + 1) % 2].append(getTriviumIndex(x - 288 + 242))
                waiting[(i + 1) % 2].append(getTriviumIndex(x - 288 + 287))
                waiting[(i + 1) % 2].append(getTriviumIndex(x - 288 + 285))
                waiting[(i + 1) % 2].append(getTriviumIndex(x - 288 + 286))
            elif x % 288 == 93:
                sys[x].append([getTriviumIndex(x - 288 - 93 + 170)])
                sys[x].append([getTriviumIndex(x - 288 - 93 + 65)])
                sys[x].append([getTriviumIndex(x - 288 - 93 + 92)])
                sys[x].append([getTriviumIndex(x - 288 - 93 + 90), getTriviumIndex(x - 288 - 93 + 91)])
                waiting[(i + 1) % 2].append(getTriviumIndex(x - 288 - 93 + 170))
                waiting[(i + 1) % 2].append(getTriviumIndex(x - 288 - 93 + 65))
                waiting[(i + 1) % 2].append(getTriviumIndex(x - 288 - 93 + 92))
                waiting[(i + 1) % 2].append(getTriviumIndex(x - 288 - 93 + 90))
                waiting[(i + 1) % 2].append(getTriviumIndex(x - 288 - 93 + 91))
            elif x % 288 == 177:
                sys[x].append([getTriviumIndex(x - 288 - 177 + 263)])
                sys[x].append([getTriviumIndex(x - 288 - 177 + 161)])
                sys[x].append([getTriviumIndex(x - 288 - 177 + 176)])
                sys[x].append([getTriviumIndex(x - 288 - 177 + 174), getTriviumIndex(x - 288 - 177 + 175)])
                waiting[(i + 1) % 2].append(getTriviumIndex(x - 288 - 177 + 263))
                waiting[(i + 1) % 2].append(getTriviumIndex(x - 288 - 177 + 161))
                waiting[(i + 1) % 2].append(getTriviumIndex(x - 288 - 177 + 176))
                waiting[(i + 1) % 2].append(getTriviumIndex(x - 288 - 177 + 174))
                waiting[(i + 1) % 2].append(getTriviumIndex(x - 288 - 177 + 175))
    
    return sys, roots, list(set(waiting[(searchRound - 1) % 2]))

def reverseSys(sys):
    rsys = {}
    for key in sys:
        for s in sys[key]:
            for x in s:
                if x in rsys:
                    rsys[x].append(key)
                else:
                    rsys[x] = [key]
    
    return rsys

def genModel(sys):
    Nvar = 0
    mapVars = {}
    for key in sys:
        for s in sys[key]:
            for x in s:
                mapVars[(key, x)] = Nvar
            Nvar = Nvar + 1
    model = Model()
    X = [i for i in range(Nvar)]
    X = model.addVars(X, lb=0.0, ub=1.0, obj=0.0, vtype=GRB.BINARY)
    
    return Nvar, mapVars, model, X

def addConstrs(sys, roots, model, mapVars, X):
    for p in sys:
        e_out = 0
        for s in sys[p]:
            e_out = e_out + X[mapVars[(p, s[0])]]
        if p in roots:
            model.addConstr(e_out == 1.0, "")
        else:
            model.addConstr(e_out <= 1.0, "")
            e_in = 0
            for x in rsys[p]:
                e_in = e_in + X[mapVars[(x, p)]]
            model.addConstr(e_in >= e_out, "")
            model.addConstr(e_in <= len(rsys[p]) * e_out, "")
    return model

def runModel(model):
    model.setParam(GRB.Param.PoolSearchMode, 2)
    model.setParam(GRB.Param.PoolSolutions, 20000000)
    #model.tune()
    model.setParam(GRB.Param.CutPasses, 10)
    model.optimize()
    return model

def analysisResult(leaves, rsys, model):
    Patterns = {}
    for s in range(model.SolCount):
        model.setParam(GRB.Param.SolutionNumber, s)
        temp = ''
        for i in leaves:
            flag = 0
            for j in rsys[i]:
                if X[mapVars[(j, i)]].Xn >= 0.5:
                    flag = 1
            temp = temp + str(flag)
        if temp in Patterns:
            Patterns[temp][0] = Patterns[temp][0] + 1
            Patterns[temp][1].append(s)
        else:
            Patterns[temp] = [1, [s]]
    
    return Patterns

def index2SupAndSub(index):
    reg = {0: "A", 93: "B", 177: "C"}
    sup = int(index / 288)
    sub = index % 288
    return [sup, sub]
    # return reg[sub]

def displayPattern(Patterns, model, sys, rsys, Nvars, mapVars, X, roots, r):
    for key in Patterns:
        if Patterns[key][0] % 2 == 0: # even patterns
            print("**********" + key + "**********")
            print("*" + " " * int((len("**********" + key + "**********") - 14) / 2) + str(Patterns[key][0]) + " sub graphs" + " " * (len("**********" + key + "**********") - 14 - int((len("**********" + key + "**********") - 14) / 2)) + "*")
            print("*" * len("**********" + key + "**********"))
            for s in Patterns[key][1]:
                model.setParam(GRB.Param.SolutionNumber, s)
                waiting = []
                for singleRoot in roots:
                    waiting.append(getTriviumIndex(singleRoot[0] * 288 + singleRoot[1]))
                while len(waiting):
                    parent = waiting.pop()
                    if parent in sys:
                        loop_short_long_double = 0
                        for child in sys[parent]:
                            if X[mapVars[(parent, child[0])]].Xn >= 0.5 and loop_short_long_double == 0:
                                print(index2SupAndSub(parent), "**>", index2SupAndSub(child[0]), end="  ")
                                waiting.append(child[0])
                                loop_short_long_double = loop_short_long_double + 1
                            elif X[mapVars[(parent, child[0])]].Xn >= 0.5 and loop_short_long_double == 1:
                                print(index2SupAndSub(parent), "-*>", index2SupAndSub(child[0]), end="  ")
                                waiting.append(child[0])
                                loop_short_long_double = loop_short_long_double + 1
                            elif X[mapVars[(parent, child[0])]].Xn >= 0.5 and loop_short_long_double == 2:
                                print(index2SupAndSub(parent), "-->", index2SupAndSub(child[0]), end="  ")
                                waiting.append(child[0])
                                loop_short_long_double = loop_short_long_double + 1
                            elif X[mapVars[(parent, child[0])]].Xn >= 0.5 and len(child) == 2:
                                print(index2SupAndSub(parent), "==>", index2SupAndSub(child[0]), end="  ")
                                print(index2SupAndSub(parent), "==>", index2SupAndSub(child[1]), end="  ")
                                waiting.append(child[0])
                                waiting.append(child[1])
                                loop_short_long_double = loop_short_long_double + 1
                            else:
                                loop_short_long_double = loop_short_long_double + 1
                                continue
                        waiting = list(set(waiting))
                print()
```
